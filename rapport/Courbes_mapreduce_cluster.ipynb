```python
import pandas as pd
import matplotlib.style as style
from scipy.interpolate import spline
from matplotlib import pyplot as plt;
from pylab import genfromtxt;  
import numpy as np
%matplotlib inline


tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]
for i in range(len(tableau20)):  
    r, g, b = tableau20[i]  
    tableau20[i] = (r / 255., g / 255., b / 255.) 

    
    

X =np.linspace(1,150,150) # taille des matrices ,           C_+ =O(n) addition de 2 nombres à n chiffres 
                                                            #C_x =O(n²) multiplication 2 nombres à n chiffres
                                                            #C_exp = O(log_2(n)) avec n l'exposant
                                                            # C_enscryption = O(log(n)^3)
                                                            # C_descryption = O(log(n)^3)
Y = 2*X**2 + (81*pow(10,-9) + 153*pow(10,-9))*X**3      # temps de calcul non sécurisé one step
Z = (1+15993*pow(10,-9))*X**2 + (81*pow(10,-9) + 609*pow(10,-9))*X**2

#Configuration du plot
plt.figure(figsize=(15,15))
ax=plt.subplot(111)
plt.style.use('fivethirtyeight')
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)  

plt.yticks(range(0, 50001, 5000), fontsize=16)  
plt.xticks(range(0, 151, 10), fontsize=16)  

for y in range(5, 50001, 5000):  
    plt.plot(range(0, 151), [y] * len(range(0, 151)), "--", lw=1, color="black", alpha=0.3)  


nomCourbes = ['version sécurisée', 'version non sécurisée']
graphNS=plt.plot(X, Y, color=tableau20[6], marker='.', label=nomCourbes[1], linewidth=1, markersize=5)
graphS=plt.plot(X, Z, color=tableau20[0], marker='.', label=nomCourbes[0], linewidth=1, markersize=5)

# Mise en place de la légende
plt.legend(prop={'size':15}, loc=(0.1,0.8))

#afficher, arranger texte

# Configuration des axes
plt.xlabel('nombre de lignes des matrices', fontsize=17); 
plt.ylabel("temps de calcul", fontsize=17)
plt.title("""Temps de calcul théorique de la multiplication de matrices en une étape\n en fonction du nombre de lignes des matrices""", fontsize=18, color=tableau20[4])
#plt.text(155, 85,"\nAuteurs: Rodrigue ABBE et Baptiste MARTINEZ"  
 #      "\nNotes: Dix mesures par taille de matrice.\nUn point par moyenne calculée\nChaque matrice est carrée.", fontsize=18, color=tableau20[10])
plt.savefig("courbesthéoriques.pdf")
plt.show()
```
